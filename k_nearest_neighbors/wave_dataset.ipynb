```python
import mglearn
import matplotlib.pyplot as plt
```

```python
x,y = mglearn.datasets.make_wave(n_samples=40)
print("x[:5]:{}".format(x[:5]))
print("y[:5]:{}".format(y[:5]))
```

```python
print("X.shape:{}".format(x.shape))
print("y.shape:{}".format(y.shape))
```

```python
plt.plot(x,y,'o')
plt.ylim(-3,3)
plt.xlabel("Feature")
plt.ylabel("Target")
```
