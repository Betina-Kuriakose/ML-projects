### loading and understanding the dataset

```python
import mglearn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
x ,y = mglearn.datasets.make_forge()
print("x[:5]:{}".format(x[:5]))
print("X.shape:{}".format(x.shape))
print("y[:5]:{}".format(y[:5]))
print("y.shape:{}".format(y.shape))
```

### visualize the data

```python
mglearn.discrete_scatter(x[:,0],x[:,1],y)
plt.legend(["Class 0","Class 1"],loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
```

### train and test 

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
print("X_train.shape:{}".format(X_train.shape))
print("X_test.shape:{}".format(X_test.shape))
print("y_train.shape:{}".format(y_train.shape))
print("y_test.shape:{}".format(y_test.shape))
```

### model creation

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn =KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

### predict

```python
print("predictions of xtest:{}",format(knn.predict(X_test)))
```

### accuracy

```python
print("test score:{}".format(knn.score(X_test,y_test)))
```

```python
fig,axes = plt.subplots(1,3,figsize=(10,3))
for n_neighbors,ax in zip([1,3,9],axes):
    clf = KNeighborsClassifier(n_neighbors=n_neighbors).fit(X_train,y_train)
    mglearn.plots.plot_2d_separator(clf,X_train,fill=True,eps=0.5,ax=ax,alpha=.4)
    mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train,ax=ax)
    mglearn.discrete_scatter(X_test[:,0],X_test[:,1],y_test,ax=ax,markeredgewidth=1)
    ax.set_title("{} neighbor(s)".format(n_neighbors))
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
axes[0].legend(loc=3)
```
